```python
%%writefile app.py
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Movie poster
st.image(
    "https://upload.wikimedia.org/wikipedia/en/b/bc/Interstellar_film_poster.jpg",
    caption="Interstellar (2014)",
    use_column_width=True
)

# Titles
st.title("Student of the AI/ML/DS")
st.header("AI")
st.subheader("DS")

# File uploader
uploadfile = st.file_uploader('Upload .csv files only', type=['csv'])

if uploadfile is not None:
    # Show file name
    file_name = uploadfile.name
    st.write(f"**Uploaded File Name:** {file_name}")

    # Read CSV
    dataFrame = pd.read_csv(uploadfile)
    st.subheader("Data uploaded successfully")
    st.success('CSV file uploaded successfully!')

    # Show columns
    columns = dataFrame.columns.tolist()
    st.write("**Columns in the file:**")
    st.write(columns)

    # Show first 5 rows
    st.write("### First 5 Rows of Your Data:")
    st.dataframe(dataFrame.head())

    # Show describe summary
    st.write("### Data Description (Summary Statistics):")
    st.dataframe(dataFrame.describe())

    # Show missing values
    st.write("### Missing Values in Each Column:")
    st.write(dataFrame.isnull().sum())

    # Correlation heatmap
    st.write("### Correlation Heatmap:")
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.heatmap(dataFrame.corr(), annot=True, fmt=".2f", cmap="coolwarm", ax=ax)
    st.pyplot(fig)

    # Histogram
    st.write("### Histogram Plot:")
    numeric_columns = dataFrame.select_dtypes(include=['float', 'int']).columns.tolist()
    if numeric_columns:
        selected_column = st.selectbox("Select a numeric column for histogram", numeric_columns)
        fig, ax = plt.subplots()
        sns.histplot(dataFrame[selected_column], kde=True, ax=ax)
        ax.set_title(f"Histogram for {selected_column}")
        st.pyplot(fig)
    else:
        st.info("No numeric columns available for histogram.")

else:
    st.error('No file uploaded yet. Please upload a CSV file to proceed.')
```
